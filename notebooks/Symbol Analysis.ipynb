```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
```

```python
def symbol_analysis(path):
    df = pd.read_pickle(path)
    df['sender_inputs'] = [[int(i) for i in j] for j in df['sender_inputs']]
    df['receiver_output'] = [[round(i, 2) for i in j] for j in df['receiver_output']]
    idx = df['sender_inputs'].apply(pd.Series).drop_duplicates().index
    new_df = df.iloc[idx]
    new_df = new_df[['sender_inputs', 'message', 'receiver_guess', 'labels',]]
    pos_df = new_df[new_df['labels'] == new_df['receiver_guess']]
    neg_df = new_df[new_df['labels'] != new_df['receiver_guess']]
    return new_df, pos_df, neg_df, df
```

```python
x, y, z, df = symbol_analysis('EGG_test/result/N10/n10v5.pkl')
grp = x.groupby('message').apply(display)
```
