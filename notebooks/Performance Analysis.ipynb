```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
```

```python
def plot_1(N, vocab_size):
    
    path = 'EGG_test/result/N'+str(N)+'/unq/output_unq'+str(N)+'_v'+str(vocab_size)+'.txt'
    title = f'Model performance over Epoch: N={N} & vocab size={vocab_size} \n unique test set'
    print(path)
    result = []
    with open(path) as f:
        for i, line in enumerate(f):
            if i<1:
                continue
            result.append(json.loads(line))
    result = pd.DataFrame(result)
    
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 3))
    for key, grp in result.groupby(['mode']):
        axes[0].plot(grp['epoch'], grp['loss'], label=key)
        axes[1].plot(grp['epoch'], grp['acc'], label=key)

    plt.suptitle(title)
    axes[0].set(xlabel="Epoch", ylabel="loss")
    axes[1].set(xlabel="Epoch", ylabel="accuracy")
    axes[0].legend()
    axes[1].legend()
    fig.tight_layout()
    plt.show()
```

```python
def plot_2(N, vocab_size):
    
    path = 'EGG_test/result/N'+str(N)+'/output'+str(N)+'_v'+str(vocab_size)+'.txt'
    path_unq = 'EGG_test/result/N'+str(N)+'/unq/output_unq'+str(N)+'_v'+str(vocab_size)+'.txt'
    title = f'Model performance over Epoch: N={N} & vocab size={vocab_size} \n unique test set'
#     title_unq = f'Model performance over Epoch: N={N} & vocab size={vocab_size}'
    print(path, path_unq)
    result = []
    with open(path) as f:
        for i, line in enumerate(f):
            if i<1:
                continue
            result.append(json.loads(line))
    result = pd.DataFrame(result)
    
    result_unq = []
    with open(path_unq) as f:
        for i, line in enumerate(f):
            if i<1:
                continue
            result_unq.append(json.loads(line))
    result_unq = pd.DataFrame(result_unq)
    result_unq = result_unq[result_unq['mode']=='test']
    result_unq['mode'] = 'test_unq'
    
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 3))
    for key, grp in result.groupby(['mode']):
        axes[0].plot(grp['epoch'], grp['loss'], label=key)
        axes[1].plot(grp['epoch'], grp['acc'], label=key)
        axes[1].axhline(0.5, linestyle='--',color='black', linewidth=.7)
    for key, grp in result_unq.groupby(['mode']):
        axes[0].plot(grp['epoch'], grp['loss'], label=key)
        axes[1].plot(grp['epoch'], grp['acc'], label=key)
    plt.suptitle(title)
    axes[0].set(xlabel="Epoch", ylabel="loss")
    axes[1].set(xlabel="Epoch", ylabel="accuracy")
    axes[0].legend()
    axes[1].legend()
    fig.tight_layout()
    plt.show()
```

```python
# 5000/1000 sample
N = 3
vocabs = [5,10,20,30,40,50]
for i in vocabs:
    plot_2(N=N, vocab_size=i) 
```
